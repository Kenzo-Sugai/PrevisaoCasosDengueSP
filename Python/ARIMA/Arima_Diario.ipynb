```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from IPython.display import display_markdown
import matplotlib.dates as mdates
```

```python
dados = pd.read_csv('.\\..\\..\\Dataset\\testes\\casos_diarios_total.csv')
#df.set_index('dt_notificacao', as_index=False, inplace=True)
```

```python
anos = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
dados_ano = {}

for ano in anos:

    dados_ano[ano] = pd.DataFrame(
                        dados \
                        .where(
                              (dados['dt_notificacao'] >= f'{ano}-01-01')
                            & (dados['dt_notificacao'] <= f'{ano}-12-31')
                        )
                    ) \
                    .dropna(subset=["dt_notificacao"])
```

```python
anos.append('TOTAL')

dados_ano['TOTAL'] = dados
```

```python
erros = {}

for ano in anos:    

    df =  dados_ano[ano]
    # Dividir os dados em treino e teste
    train_size = int(len(df) * 0.8)
    train, test = df.iloc[:train_size], df.iloc[train_size:]

    # Treinar o modelo ARIMA
    p, d, q = 60, 1, 0 
    # p: dias passados para analise
    # d: quantidade de dias
    # q: usar media movel (booleano)
    model = ARIMA(train['casos'], order=(p, d, q))
    model_fit = model.fit()

    # Fazer previses
    forecast = model_fit.forecast(steps=len(test))

    # Avaliar o modelo
    erros[ano] = {
        'mse': mean_squared_error(test['casos'], forecast),
        'rmse': root_mean_squared_error(test['casos'], forecast), 
        'mae':mean_absolute_error(test['casos'], forecast),
        'mape': mean_absolute_percentage_error(test['casos'], forecast)
    }

    display_markdown(f'### {ano}', raw=True)

    # Plotar os resultados
    plt.figure(figsize=(14, 8), dpi=100)
    plt.plot(train['dt_notificacao'], train['casos'], label='Treino')
    plt.plot(test['dt_notificacao'], test['casos'], label='Teste', color='orange')
    plt.plot(test['dt_notificacao'], forecast, label='Previso ARIMA', color='green')
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(bymonthday=1))
    plt.title('Previso com ARIMA')
    plt.xlabel('Data')
    plt.ylabel('Casos')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
for ano in anos:
    print(ano)
    print(erros[ano])
```
